### Populära typer

```python
types = pd.concat([df["Type 1"],df["Type 2"]])
common_types = types.value_counts()
```

```python
color_map = {
    'Water': 'blue', "Psychic": "#eb49f9", "Fire": "red", "Grass": "green", "Electric": "yellow", "Normal": "#c6c4c6",
    "Fighting": "orange", "Flying": "#01d2d8", "Poison": "purple", "Ground": "brown", "Rock": "#d1cc94", "Ice": "#1efef7",
    "Bug": "#82c203", "Dragon": "#0c03c2", "Ghost": "#74078c", "Dark": "#260233", "Steel": "#c3dee8", "fairy": "#fca8fa"}
colors = [color_map[types] for types in common_types.index[:10]]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=common_types.index[:10], y=common_types.values[:10], palette=colors)
plt.title("de 10 mest populära typerna")
plt.ylabel("mängd pokemon med typen")
plt.xlabel("typ")
plt.xticks(rotation=45)
plt.show
```
